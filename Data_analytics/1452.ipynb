```python
def calc(x, temp):
#     time from x to x +1
    group = temp.groupby(['VehicleJourneyID', 'TimeFrame'])
    average = []
    if x == len(stops_145)-1:
        return 0
    for name, grou in group:
        start = grou['Timestamp'].loc[grou['StopID'] == stops_145[x]].mean()
        stop = grou['Timestamp'].loc[grou['StopID'] == stops_145[x+1]].mean()
        average += [stop - start]
    average = [value for value in average if not math.isnan(value)]
    if len(average) != 0:
        answer = sum(average)/len(average)
        #answer is in micro seconds
        answer = answer *(10**-6)
        #answer is now in seconds
        return answer
    else:
        return 0
```
