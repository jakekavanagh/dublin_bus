```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.font_manager as fm
import matplotlib
import pickle
import timeit
import h5py
import dask.dataframe as dd
%matplotlib inline
```

```python
res = pd.read_csv('siri.20130114.csv', header=None, low_memory= False)
res.columns = pd.Series(['Timestamp', 'LineID', 'Direction', 'JourneyPatternID', 'TimeFrame', 'VehicleJourneyID', 'Operator', 'Congestion', 'Lon', 'Lat', 'Delay', 'BlockID', 'VehicleID', 'StopID', 'AtStop'])
res['Timestamp'] = pd.to_datetime(res['Timestamp'], unit='us')
```

```python
res['LineID'] = res['LineID'].astype('object')
res["LineID"].fillna("null", inplace=True)
res["JourneyPatternID"].fillna("null", inplace=True)
```

```python
def f(x):
    if x[0] != 'null':
        return str(x[0][:4]).lstrip('0')
    else:
        return x[1]

df = pd.concat([res['JourneyPatternID'], res['LineID']], axis = 1)
df['RouteII'] = df.apply(f, axis=1)
```

```python
res = pd.concat([res,df['RouteII']], axis=1)
res['Timestamp'] = res['Timestamp'].map(lambda x: str(x))
res = res.sort_values('Timestamp')
res
```
